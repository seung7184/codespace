```python
def timeConversion(s):
    ans = s[0:8]
    spl = ans.split(':')
    if s[8:] == 'AM':
        if spl[0] == '12':
            spl[0] = int(spl[0])-12
            return f'{spl[0]}0:{spl[1]}:{spl[2]}'
        else:
            return ans
    else:
        if spl[0] == '12':
            return f'{spl[0]}:{spl[1]}:{spl[2]}'
        else:
            spl[0] = int(spl[0])+12
            return f'{spl[0]}:{spl[1]}:{spl[2]}'
```

```python
def findZigZagSequence(a, n):
    a.sort()
    mid = int((n + 1)/2)
    a[mid-1], a[n-1] = a[n-1], a[mid-1]
    print(a)

    st = mid 
    ed = n - 1 
    while(st <= ed and st < n and ed < n):
        a[st], a[ed-1] = a[ed-1], a[st]
        st = st + 1
        ed = ed + 1

    for i in range (n):
        if i == n-1:
            print(a[i])
        else:
            print(a[i], end = ' ')
    return

test_cases = 1
for cs in range (test_cases):
    n = 7
    a = list(map(int, input().split()))
    findZigZagSequence(a, n)
```
